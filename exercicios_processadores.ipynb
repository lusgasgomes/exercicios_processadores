    ADD Op. - 2 Ciclos, 1 para ler o endereço e outro para ler o dado armazenado no endereço decodificado, incrementando R0.

    SUB Op. - 2 Ciclos, 1 para ler o endereço, outro para ler o dado, decrementando com o valor de R0.  

    ADD Op.1, Op.2 - 4 ciclos, dessas, 3 para leitura. 1 para ler a instrução, outra para ler o dado do Op.1, outro para ler o dado do Op.2 e somar com o Op.1, e por ultimo, 1 para escrever a soma dos dois operandos na célula equivalente ao op.1

    INCR - 1 ciclo, incremento de R0.

    LDA - 2 ciclos, 1 para leitura do endereço e outra paraa leitura do operando que será adicionado em R0.
